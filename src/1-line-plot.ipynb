### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11, 1)
y = x ** 2

plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Y = X²')

plt.title('Square Function Plot')
plt.xlabel('X Values')
plt.ylabel('Y = X²')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.show()
```
